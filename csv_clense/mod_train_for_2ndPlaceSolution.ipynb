- delete_list_not_similar.csv
- train.csv
- image_assign_group.csv

この3種類を使用

1. train.csvを複製、my_train_w_group_labelを作成
2. my_train_w_group_labelから、delete_list_not_similar.csvのgid列を持つものを削除
3. train.csvのgidに対応するimage_assign_group.csvのsame_groupを、新しく[label]列として定義、グループ化

```python
import pandas as pd

# CSVファイルを読み込む
df_train = pd.read_csv('train.csv')
df_delete_list = pd.read_csv('delete_list_not_similar.csv')
df_image_assign_group = pd.read_csv('image_assign_group.csv')

# train.csvを複製してmy_train_w_group_labelを作成
df_my_train_w_group_label = df_train.copy()

# delete_list_not_similar.csvに記載されているgidを持つ行を削除
df_my_train_w_group_label = df_my_train_w_group_label[~df_my_train_w_group_label['gid'].isin(df_delete_list['del_gid'])]

# gidに対応するimage_assign_group.csvのsame_groupを新しいlabel列として割り当て
df_my_train_w_group_label = df_my_train_w_group_label.merge(df_image_assign_group[['gid', 'same_group']], on='gid')
df_my_train_w_group_label = df_my_train_w_group_label.rename(columns={'same_group': 'same_group'})

# 新しいCSVファイルとして出力
df_my_train_w_group_label.to_csv('my_train_w_group_label.csv', index=False)
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# 既存のCSVファイルを読み込む
df = pd.read_csv('my_train_w_group_label.csv')

# LabelEncoderのインスタンスを作成
label_encoder = LabelEncoder()

# label列をエンコード
df['label'] = label_encoder.fit_transform(df['same_group'])

# 新しいCSVとして保存
df.to_csv('my_train_w_group_label.csv', index=False)
```
